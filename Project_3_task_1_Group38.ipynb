```python
#importing necessary libraries for keyword networks graphical representation
import matplotlib.pyplot as plt
import networkx as nx
V = nx.from_pandas_adjacency(df_adj_matrix)
```

```python
# to find the Degree of the graph
V.degree
```

```python
V.degree(weight='weight')
```

```python
V.edges()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
#plotting keyword networks
plt.figure(figsize=(40,25))
graph_pos = nx.spring_layout(V)
graph_edges = V.edges()
nx.draw_networkx_nodes(V, graph_pos, node_size=25, node_color='red', alpha=0.6)
nx.draw_networkx_edges(V, graph_pos,width=0.5)
nx.draw_networkx_labels(V, graph_pos, font_size=7, font_family='sans-serif',font_color='red')

plt.show()
```

```python
#Strength Calculation - 
# Strength can be calculated by summing up the weights row-wise or column wise
count = 0
df_result_set = pd.DataFrame(columns = ['Node Name','Strength','Degree'])
for i in df_adj_matrix.columns :
  degree = (df_adj_matrix[i] != 0).sum()
  df_result_set = df_result_set.append({'Node Name': i,'Strength': df_adj_matrix[i].sum(),'Degree': degree}, ignore_index=True)
```

```python
df_result_set
```

```python
#displaying top 10 nodes by strength
df_result_set[['Node Name','Strength']].sort_values(by=['Strength'],ascending=False).head(10)
```

```python
#displaying the top 10 nodes by degree
df_result_set[['Node Name','Degree']].sort_values(by=['Degree'],ascending=False).head(10)
```

Now, we need to display top 10 node pairs by weight

```python
df_weights=pd.DataFrame(columns=['Node 1','Node 2', 'Weight'])
```

```python
#appending the dataset with values
for i in df_adj_matrix.index:
  for j in df_adj_matrix.columns:
    df_weights = df_weights.append({'Node 1': i,'Node 2': j,'Weight': df_adj_matrix[i][j]}, ignore_index=True)
```

```python
# Top 10 Node Pairs by Weight
df_weights.sort_values(by=['Weight'],ascending=False).head(10)
```
